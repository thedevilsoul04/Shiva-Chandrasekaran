```python
import numpy as np
import pandas as pd
data = pd.read_csv('/content/Popular_Spotify_Songs.csv', encoding='latin-1')
data.head()
```

```python
data.tail()
```

```python
data.isnull()
data.isnull().sum()
```

```python
obj_list=data.select_dtypes("object").columns
print(obj_list)
```

```python
for obj in obj_list:
  print(obj)
```

```python
from sklearn.preprocessing import LabelEncoder
for obj in obj_list:
 data[obj]=LabelEncoder().fit_transform(data[obj])
 data
```

```python
X=data.drop('in_spotify_playlists',axis=1)
Y=data['in_spotify_playlists']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
X_train
X_test
Y_test
```

```python
from sklearn.ensemble import RandomForestRegressor #random forest
Model=RandomForestRegressor()
Model.fit(X_train,Y_train)
Model.score(X_test,Y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier #decision tree
thala = DecisionTreeClassifier()
thala.fit(X_train, Y_train)
thala.score(X_test, Y_test)
```

```python
from sklearn.svm import SVC  #svc
ak=SVC()
ak.fit(X_train,Y_train)
ak.predict(X_test)
ak.score(X_test,Y_test)*100
```
